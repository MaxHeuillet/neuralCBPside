```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
import multiprocessing

# Define your neural network model
class MyModel(nn.Module):
    def __init__(self):
        super(MyModel, self).__init__()
        self.fc1 = nn.Linear(100, 50)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(50, 50)
        self.relu = nn.ReLU()
        self.fc3 = nn.Linear(50, 5)
        self.relu = nn.ReLU()
        self.fc4 = nn.Linear(5, 2)

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.relu(x)
        x = self.fc3(x)
        x = self.relu(x)
        x = self.fc4(x)
        return x

# Function to train a single model
def train_model(args):
    model, dataloader = args
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.SGD(model.parameters(), lr=0.01)

    model.to('cuda')

    for epoch in range(10):  # Replace with the number of epochs you want to run
        print(epoch)
        for inputs, labels in dataloader:
            inputs, labels = inputs.to('cuda'), labels.to('cuda')
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

# Create an instance of the model
model1 = MyModel()
model2 = MyModel()
model3 = MyModel()
model4 = MyModel()

# Define your dataset and dataloader
# Replace this with your actual dataset and DataLoader setup
# For simplicity, we're using random data here
dummy_data = torch.randn(4000, 100)
dummy_labels = torch.randint(0, 2, (4000,))
dataset = torch.utils.data.TensorDataset(dummy_data, dummy_labels)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)

# Combine models and dataloaders into a list of tuples
models = [model1, model2, model3, model4]
datas = [dataloader] * len(models)  # Create a list of tuples

pool = multiprocessing.Pool(processes=4)
pool.map( train_model, zip(models, datas) )


print("Training of multiple models complete!")
```
